# Experiment with other types of charts

```python
#concept of joins
#create practise data sets
df1 = pd.DataFrame({
   "employee" : ["ABC","XYZ","MNO"],
   "age" : ["22","25","30"]    
})

df1
```

```python
df2 = pd.DataFrame({
   "employee" : ["ABC","XYZ","PQR"],
   "salary" : ["10000","125000","30000"]    
})

df2
```

```python
#inner join
df3 = pd.merge(df1, df2, on = 'employee')
df3
```

```python
#outer join
df3 = pd.merge(df1, df2, on = 'employee', how = "outer")
df3
```

```python
#left join
df3 = pd.merge(df1, df2, on = 'employee', how = "left")
df3
```

```python
#right join
df3 = pd.merge(df1, df2, on = 'employee', how = "right")
df3
```
